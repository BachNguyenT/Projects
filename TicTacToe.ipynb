```python
import random

# generate a spot for the bot to place by generating a random number in range of game board
def bot(num):
    randomNum = random.choice(num)
    return randomNum

# get player's input of location to place the piece
def askPlayer(availableNum):
    # playerNum=1000
    # while playerNum not in (availableNum):
    playerNum = int(input('Input a location: '))
    while playerNum not in availableNum:
        playerNum = int(input('Input denied, please choose a legal spot: '))
    return playerNum

# check if a win happened, returns ture or false
def checkWin(num):
    win = False

    for i in range(3):
        if num[i] == num[i + 3] == num[i + 6] == 'O' or num[i] == num[i + 3] == num[i + 6] == 'X':
            win = True
        
    if num[0] == num[1] == num[2] == 'O' or num[0] == num[1] == num[2] == 'X':
        win = True
    elif num[3] == num[4] == num[5] == 'O' or num[3] == num[4] == num[5] == 'X':
        win = True
    elif num[6] == num[7] == num[8] == 'O' or num[6] == num[7] == num[8] == 'X':
        win = True
    elif num[0] == num[4] == num[8] == 'O' or num[0] == num[4] == num[8] == 'X':
        win = True
    elif num[2] == num[4] == num[6] == 'O' or num[2] == num[4] == num[6] == 'X':
        win = True
    return win


# declare variables
num = [1,2,3,4,5,6,7,8,9]
availableNum = [1,2,3,4,5,6,7,8,9]
# gameBoard = (' ' + str(num[0]) + ' | ' + str(num[1]) + ' | ' + str(num[2]) + ' \n-----------\n ' +  str(num[3]) + ' | ' + str(num[4]) +  ' | ' + str(num[5]) + ' \n-----------\n '+ str(num[6])+ ' | '+ str(num[7]) + ' | ' + str(num[8]) + ' ')
win = False
roundCounter = 1


print ((' ' + str(num[0]) + ' | ' + str(num[1]) + ' | ' + str(num[2]) + ' \n-----------\n ' +  str(num[3]) + ' | ' + str(num[4]) +  ' | ' + str(num[5]) + ' \n-----------\n ' + str(num[6])+ ' | '+ str(num[7]) + ' | ' + str(num[8]) + ' '))

while win == False:
    # player
    if roundCounter % 2 == 1:
        playerNum = askPlayer(availableNum)
        num[playerNum-1] = 'O'
        availableNum.remove(playerNum)

        print ((' ' + str(num[0]) + ' | ' + str(num[1]) + ' | ' + str(num[2]) + ' \n-----------\n ' +  str(num[3]) + ' | ' + str(num[4]) +  ' | ' + str(num[5]) + ' \n-----------\n ' + str(num[6])+ ' | '+ str(num[7]) + ' | ' + str(num[8]) + ' '))

        win = checkWin(num)
        if win == True:
            print ("You Win!")
            break

    # bot
    if roundCounter % 2 == 0:
        randomNum = bot(availableNum)
        num[randomNum-1] = 'X'
        availableNum.remove(randomNum)

        print ((' ' + str(num[0]) + ' | ' + str(num[1]) + ' | ' + str(num[2]) + ' \n-----------\n ' +  str(num[3]) + ' | ' + str(num[4]) +  ' | ' + str(num[5]) + ' \n-----------\n ' + str(num[6])+ ' | '+ str(num[7]) + ' | ' + str(num[8]) + ' '))

        win = checkWin(num)
        if win == True:
            print ("You Lose!")
            break

    roundCounter +=1
    if roundCounter == 10:
        print ("You Tie!")
        break
    
```
